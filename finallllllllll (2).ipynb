```python
 #create the new DataFrame with columns 'year', 'popular_genre', and 'number of release'
popular_genres = pd.DataFrame(columns=['year', 'popular_genre', 'number of release'])

# iterate through each unique year in the DataFrame
for year in newdf['year'].unique():
    if year not in [2025, 2024, 2023]:  # exclude years 2025, 2024, and 2023
        df_year = newdf[newdf['year'] == year]  # filter by year
        if not df_year.empty:  # check if the resulting DataFrame is not empty
            # count the occurrences of each genre in the year
            genre_counts = df_year['genre'].value_counts()
            # get the most frequent genre
            popular_genre = genre_counts.index[0]
            # get the count of the popular genre
            count = genre_counts.iloc[0]
            # add the year, popular_genre, and count to the new DataFrame
            popular_genres = pd.concat([popular_genres, pd.DataFrame({'year': [year], 'popular_genre': [popular_genre], 'number_of_release': [count]})], ignore_index=True)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
X = newdf['rating'].values.reshape(-1, 1)
y = newdf['votes'].values.reshape(-1, 1)
model = LinearRegression()
model.fit(X, y)
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='Data')
plt.plot(X, model.predict(X), color='red', label='Linear Regression')

plt.title('Linear Regression of Movie ratings vs votes')
plt.xlabel('Rating')
plt.ylabel('Votes')
plt.legend()
plt.grid(True)
plt.show()
```

```python
data_genres = newdf.assign(genre=df['genre'].str.split(',')).explode('genre')
data_genres = data_genres.reset_index()[['index', 'genre']]
data_genres.columns = ['index', 'genre']
data_genres = data_genres.set_index('index')
data_genres.head(5)
```

```python
genre_matrix = pd.crosstab(index=data_genres.index, columns=data_genres['genre'])
genre_matrix
```

```python
genre_coocurrence = genre_matrix.T.dot(genre_matrix)
genre_coocurrence
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
sns.heatmap(genre_coocurrence, cmap='YlGnBu', annot=True, fmt='.0f')
plt.title('Movie Genre Co-occurrence')
plt.show()
```

```python
np.fill_diagonal(genre_coocurrence.values, 0)
sns.heatmap( genre_coocurrence )
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
vec = CountVectorizer()

vec_matrix = vec.fit_transform(newdf['genre'])
similarity = cosine_similarity(vec_matrix)
```

```python
def recommend_movie(movie):
    if movie not in newdf['movie_name'].unique():
        return('Sorry! The movie you requested is not in our database. Please check the spelling or try with some other movies')
    else:
        i = newdf.loc[newdf['movie_name']==movie].index[0]
        lst = list(enumerate(similarity[i]))
        lst = sorted(lst, key = lambda x:x[1] ,reverse=True)
        lst = lst[1:11] # excluding first item since it is the requested movie itself
        l = []
        year=[]
        for i in range(len(lst)):
            a = lst[i][0]
            l.append(newdf['movie_name'][a])
            year.append(newdf['year'][a])
            
        
        df2 = pd.DataFrame({'Movies Recommended':l, 'Year':year})
        df2.drop_duplicates
        return df2
```
